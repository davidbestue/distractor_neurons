```python
time_s =7
w=10 #100ms
f = np.shape(Matrix_spikes_on)[1] #700
t1s = np.arange(0,f,w)
t2s = np.arange(w,f+w,w)
```

```python
fr_time = []

for N in range(np.shape(Matrix_spikes_on)[0]):
    neuron_fr = []
    for i in range(len(t1s)):
        neuron_fr.append(Matrix_spikes_on[N, t1s[i]:t2s[i]].sum()/ (w*time_s/f) )
    #
    fr_time.append(neuron_fr)
```
