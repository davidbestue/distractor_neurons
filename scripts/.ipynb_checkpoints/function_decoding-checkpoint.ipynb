### Group trials by condition (Target loc)

```python
Conditions={}
Conditions['control_LIP']=[]
Conditions['0_LIP']=[]
Conditions['100_LIP']=[]
Conditions['200_LIP']=[]
Conditions['800_LIP']=[]

Conditions['control_DLPFC']=[]
Conditions['0_DLPFC']=[]
Conditions['100_DLPFC']=[]
Conditions['200_DLPFC']=[]
Conditions['800_DLPFC']=[]


for file_to_use in os.listdir():
    name = file_to_use.split('.')[0]
    dict_n = np.load(file_to_use).item()
    
    ### Add other columns about the target presentation
    
    ### the target must go from 0,180
    Centered_target = [circdist(dict_n['Descriptor'].loc[27][i], dict_n['Descriptor'].loc[28][i]) for i in range(0, len(dict_n['Descriptor'].loc[28]))]
    dict_n['Descriptor'].loc[30]=Centered_target  ##### 0, 180 target presentation, RF at 0
    
    target_180_180 = dict_n['Descriptor'].loc[28] - dict_n['Descriptor'].loc[27]
    for n in range(0, len(target_180_180)):
        if target_180_180[n] >180:
            target_180_180[n]= target_180_180[n]-360
        if target_180_180[n] < -180:
            target_180_180[n]= target_180_180[n] + 360 
    
    dict_n['Descriptor'].loc[31]=target_180_180  ##### -180, 180 target presentation, RF at 0
    
    target_0_360_centered = dict_n['Descriptor'].loc[31]
    
    for n in range(0, len(target_0_360_centered)):
        if target_0_360_centered[n] ==-45:
            target_0_360_centered[n]= 315
        if target_0_360_centered[n] == -90:
            target_0_360_centered[n]= 270
        if target_0_360_centered[n] == -135:
            target_0_360_centered[n]= 225
        if target_0_360_centered[n] == -180:
            target_0_360_centered[n]= 180
            
    
    dict_n['Descriptor'].loc[32]=target_0_360_centered  ##### 0, 360 target presentation, RF at 0
    dict_n['Descriptor'].loc[33]=name  ##### 0, 360 target presentation, RF at 0
    
    #column of the response????
    Response_locations = dict_n['Descriptor'].loc[18].values
    rf_loc = dict_n['Descriptor'].loc[27].values[0]
    Response_loc = [angle_from_RF(r, rf_loc) for r in Response_locations]
    dict_n['Descriptor'].loc[34]=Response_loc #### the real angle of the response (0, 45, 90...)
    
    
    
    
    #
    B_A = brain_area(name)
    ###
    for DELAY in [0, 100, 200, 800]:
        #TYPE = dict_n['Descriptor'].loc[24]
        #
        if DELAY==0:
            Trials = dict_n['Descriptor'].columns[(dict_n['Descriptor'].loc[20]==DELAY) & (dict_n['Descriptor'].loc[19]==99)]
            beh = dict_n['Descriptor'][Trials].loc[28] ## append behaviour
            neuron = dict_n['Descriptor'][Trials].loc[33] ## append neuron   
            firing = dict_n['data_norm'].loc[Trials].iloc[:, :22] ## append Firing rates
            FR = pd.concat([beh, firing, neuron], axis=1)
            Conditions['control_' + B_A].append(FR)
            #########
            Trials = dict_n['Descriptor'].columns[(dict_n['Descriptor'].loc[20]==DELAY) & (dict_n['Descriptor'].loc[24]!=0)]
            beh = dict_n['Descriptor'][Trials].loc[28] ## append behaviour
            neuron = dict_n['Descriptor'][Trials].loc[33] ## append neuron   
            firing = dict_n['data_norm'].loc[Trials].iloc[:, :22] ## append Firing rates
            FR = pd.concat([beh, firing, neuron], axis=1)
            Conditions[str(DELAY) + '_'  + B_A].append(FR)
        else:
            Trials = dict_n['Descriptor'].columns[(dict_n['Descriptor'].loc[20]==DELAY) & (dict_n['Descriptor'].loc[24]!=0)]
            beh = dict_n['Descriptor'][Trials].loc[28] ## append behaviour
            neuron = dict_n['Descriptor'][Trials].loc[33] ## append neuron   
            firing = dict_n['data_norm'].loc[Trials].iloc[:, :22] ## append Firing rates
            FR = pd.concat([beh, firing, neuron], axis=1)
            Conditions[str(DELAY) + '_'  + B_A].append(FR)
            

            
#### Concatenate neurona and plot the FR at each time point
## The FR is normalized

## Firing rates all conditions
LIP_control = pd.concat(Conditions['control_LIP'])
LIP_0 = pd.concat(Conditions['0_LIP'])
LIP_100 = pd.concat(Conditions['100_LIP'])
LIP_200 = pd.concat(Conditions['200_LIP'])
LIP_800 = pd.concat(Conditions['800_LIP'])

DLPFC_control = pd.concat(Conditions['control_DLPFC'])
DLPFC_0 = pd.concat(Conditions['0_DLPFC'])
DLPFC_100 = pd.concat(Conditions['100_DLPFC'])
DLPFC_200 = pd.concat(Conditions['200_DLPFC'])
DLPFC_800 = pd.concat(Conditions['800_DLPFC'])

plt.title('delay LIP')
sns.pointplot(data=LIP_control.iloc[:, 1:-1], color='blue')
sns.pointplot(data=LIP_0.iloc[:, 1:-1], color='green')
sns.pointplot(data=LIP_100.iloc[:, 1:-1], color='orange')
sns.pointplot(data=LIP_200.iloc[:, 1:-1], color='black')
sns.pointplot(data=LIP_800.iloc[:, 1:-1], color='red')
plt.show()

plt.title('delay DLPFC')
sns.pointplot(data=DLPFC_control.iloc[:, 1:-1], color='blue')
sns.pointplot(data=DLPFC_0.iloc[:, 1:-1], color='green')
sns.pointplot(data=DLPFC_100.iloc[:, 1:-1], color='orange')
sns.pointplot(data=DLPFC_200.iloc[:, 1:-1], color='black')
sns.pointplot(data=DLPFC_800.iloc[:, 1:-1], color='red')
plt.show()
            
```

```python

def decoding_spikes_angles_err(df, size_test=0.1):
    #### Input : dataframe with three columns: (spikes, behaviour and beuron label)
    ###Take off nans
    df = df.loc[df.iloc[:,0]<9999]
    df.columns=['spikes', 'beh', 'neuron']
    neuron_means=[]
    ### Train and test on the same neuron --> append the error of each neuron
    pop_err=[]
    
    for Neur in df.neuron.unique():
        ### same neuron 5 times --> mean value
        neur_err=[]
        for rep in range(0,5):
            spikes_train, spikes_test, beh_train, beh_test = train_test_split(df.groupby('neuron').get_group(Neur)['spikes'],
                                                                              df.groupby('neuron').get_group(Neur)['beh'],
                                                                              test_size=size_test)  

            ######## Trainning #########
            ## X matrix (intercept and spikes)
            X = np.column_stack([np.ones(np.shape(spikes_train)[0]),spikes_train])
            ## Y (sinus and cos)
            sinus =np.sin([np.radians(np.array(beh_train)[i]) for i in range(0, len(beh_train))])
            cosinus = np.cos([np.radians(np.array(beh_train)[i]) for i in range(0, len(beh_train))])
            Y = np.column_stack([cosinus, sinus])
            ### one OLS for sin and cos: output: beta of intercetp and bea of spikes (two B intercepts and 2 B for spikes )
            model = sm.OLS(Y, X)
            ##train the model
            fit=model.fit()

            ######### Testing ###########
            X = np.column_stack([np.ones(np.shape(spikes_test)[0]),spikes_test])
            p = fit.predict(X)
            x = p[:,0]
            y = p[:,1]
            #####
            ##### Error --> take the resulting vector in sin/cos space
            ### from sin and cos get the angle (-pi, pi)
            #pred_angle = [ np.degrees(np.arctan2(y[i], x[i]) + np.pi) for i in range(0, len(y))]
            pred_angle = [ np.degrees(np.arctan2(y[i], x[i])) for i in range(0, len(y))]
            for i in range(0, len(pred_angle)):
                if pred_angle[i]<0:
                    pred_angle[i]=360+pred_angle[i]
            ##
            error=[ circdist(beh_test[i], pred_angle[i]) for i in range(0, len(pred_angle))]

            #low_value --> predicted positionns close to real
            neur_err.append(np.mean(error))
            ####
        
        #####       
        pop_err.append(np.mean(neur_err))
        ####
    
    return np.mean(pop_err)
```
